```python
import backtrader as bt
from datetime import time
import pandas as pd

from indicators import *
from functions import *


'''default strat structure to easily copy and paste'''
# class strat(bt.Strategy):

#     def __init__(self,df):
        # self.df = df.reset_index()

        # self.sl_percent = 5
        # self.tp_percent = 5

        # stoch_rsi(self.df)
        # hull(self.df)
        # qqe(self.df)
        # atr(self.df)



#     def next(self):
#         index = len(self)-1 #get index of current candle to get df values


#         '''if not currently in the market'''
#         if not self.position:

#             '''buy conditions'''
#             if :

            
#             '''sell conditions'''
#             if :



#         '''if already in a position'''
#         if self.position:


#             '''check long'''
            # if self.position.size > 0:


                # #stop loss
                # if self.data.low < self.position.price - (self.position.price*(self.sl_percent/100)):
                #     self.close()

                # #take profit
                # if self.data.high > self.position.price + (self.position.price*(self.tp_percent/100)):
                #     self.close()


#             '''check short'''
            # if self.position.size < 0:

                # #stop loss
                # if self.data.high > self.position.price + (self.position.price*(self.sl_percent/100)):
                #     self.close()

                # #take profit
                # if self.data.low < self.position.price - (self.position.price*(self.tp_percent/100)):
                #     self.close()




class Strat1(bt.Strategy):

    def __init__(self,df):
        self.df = df.reset_index()

        self.sl_percent = 5
        self.tp_percent = 5

        stoch_rsi(self.df)
        hull(self.df)
        qqe(self.df)
        atr(self.df)



    def next(self):
        index = len(self)-1 #get index of current candle to get df values

        True_Range = self.df['True_Range'].iloc[index]
        hull = self.df['hull'].iloc[index]

        '''if not currently in the market'''
        if not self.position:

            '''buy conditions'''
            if (True_Range >= 30 and True_Range >= 44) and (hull >= 2443 and hull <= 3427):
                self.buy()
            
            '''sell conditions'''
            if (True_Range >= 30 and True_Range <= 44) and (hull >= 2451 and hull <= 3435):
                self.sell()


        '''if already in a position'''
        if self.position:


            '''check long'''
            if self.position.size > 0:
                if (True_Range >= 30 and True_Range <= 44) and (hull >= 2451 and hull <= 3435):
                    self.close()


                #stop loss
                if self.data.low < self.position.price - (self.position.price*(self.sl_percent/100)):
                    self.close()

                #take profit
                if self.data.high > self.position.price + (self.position.price*(self.tp_percent/100)):
                    self.close()


            '''check short'''
            if self.position.size < 0:
                if (True_Range >= 30 and True_Range >= 44) and (hull >= 2443 and hull <= 3427):
                    self.close()

                #stop loss
                if self.data.high > self.position.price + (self.position.price*(self.sl_percent/100)):
                    self.close()

                #take profit
                if self.data.low < self.position.price - (self.position.price*(self.tp_percent/100)):
                    self.close()





class Gulf(bt.Strategy):

    def __init__(self,df):
        self.df = df.reset_index()

        self.sl_percent = 5
        self.tp_percent = 5

        stoch_rsi(self.df)
        hull(self.df)
        qqe(self.df)
        atr(self.df)



    def next(self):
        index = len(self)-1 #get index of current candle to get df values
        
        stoch_rsi = self.df['stoch_rsi'].iloc[index]
        true_range = self.df['True_Range'].iloc[index]

        '''if not currently in the market'''
        if not self.position:

            '''buy conditions'''
            if stoch_rsi > .6 and true_range <= 60:
                self.buy()

            
            '''sell conditions'''
            if stoch_rsi <= .4 and true_range <= 55:
                self.sell()



        '''if already in a position'''
        if self.position:


            '''check long'''
            if self.position.size > 0:
                if stoch_rsi <= .4 and true_range <= 55:
                    self.close()
                
                
                #stop loss
                if self.data.low < self.position.price - (self.position.price*(self.sl_percent/100)):
                    self.close()

                #take profit
                if self.data.high > self.position.price + (self.position.price*(self.tp_percent/100)):
                    self.close()



            '''check short'''
            if self.position.size < 0:
                if stoch_rsi > .6 and true_range <= 60:
                    self.close()
                
                
                #stop loss
                if self.data.high > self.position.price + (self.position.price*(self.sl_percent/100)):
                    self.close()

                #take profit
                if self.data.low < self.position.price - (self.position.price*(self.tp_percent/100)):
                    self.close()



class Pop(bt.Strategy):

    def __init__(self,df):
        self.df = df
        self.tp_percent = .3
        self.sl_percent = .25
        self.consolidation_percent = 1

        atr(self.df)

        self.df = self.df.reset_index()



    def next(self):
        index = len(self)-1 #get index of current candle to get df values
        

        atr = self.df['True_Range'].iloc[index]

        tr_deviance = (atr/self.data.close[0])*100


        '''if not currently in the market'''
        if not self.position:

            #check if in a time of little movement
            if tr_deviance < self.consolidation_percent:

                low,high = get_extreme_of_consolidation(self.df.loc[:index],self.consolidation_percent) #get low and high price of current consolidation

                '''buy conditions'''
                if self.data.close <= ((high-low)/3)+low:
                    self.buy()
                
                '''sell conditions'''
                if self.data.close >= high-((high-low)/3):
                    self.sell()



        '''if already in a position'''
        if self.position:


            '''check long'''
            if self.position.size > 0:

                #stop loss
                if self.data.low < self.position.price - (self.position.price*(self.sl_percent/100)):
                    self.close()

                #take profit
                if self.data.high > self.position.price + (self.position.price*(self.tp_percent/100)):
                    self.close()


            '''check short'''
            if self.position.size < 0:

                #stop loss
                if self.data.high > self.position.price + (self.position.price*(self.sl_percent/100)):
                    self.close()

                #take profit
                if self.data.low < self.position.price - (self.position.price*(self.tp_percent/100)):
                    self.close()






class PineStrat(bt.Strategy):

    def __init__(self,df):
        self.df = df.reset_index() #reset index of df for use by indicators

        self.tp_risk_percent = 1.50

        self.hull = hull(self.df)
        self.qqe = qqe(self.df)
        self.volOsc = calc_vol_osc(self.df)

        self.qqe['green_cross'] = (self.qqe['greenbar1'] > self.qqe['greenbar2']) & (self.qqe['greenbar1'].shift(1) < self.qqe['greenbar2'].shift(1))
        self.qqe['red_cross'] = (self.qqe['redbar1'] > self.qqe['redbar2']) & (self.qqe['redbar1'].shift(1) < self.qqe['redbar2'].shift(1))




    def next(self):

        index = len(self)-1

        '''if not currently in the market'''
        if not self.position:

            '''buy conditions'''
            if self.qqe['green_cross'].iloc[index] and self.hull['hull_green'].iloc[index] and self.hull['closeover'].iloc[index] and self.volOsc['VolOsc_over_0'].iloc[index]:
                self.buy()


            '''sell conditions'''
            if self.qqe['red_cross'].iloc[index] and self.hull['hull_red'].iloc[index] and self.hull['closeunder'].iloc[index] and self.volOsc['VolOsc_under_0'].iloc[index]:
                self.sell()


        '''if already in a position'''
        if self.position:

            '''check long'''
            if self.position.size > 0:

                #stop loss
                stop = min(self.data.low.get(ago=0,size=10))
                if self.data.low < stop:
                    self.close(price=stop)

                #take profit
                profit = self.data.close + abs((self.data.close - stop)*self.tp_risk_percent)
                if self.data.high > profit:
                    self.close(price=profit)


            '''check short'''
            if self.position.size < 0:

                #stop loss
                stop = max(self.data.high.get(ago=0,size=10))
                if self.data.high > stop:
                    self.close(price=stop)

                #take profit
                profit = self.data.close - abs((self.data.close - stop)*self.tp_risk_percent)
                if self.data.low < profit:
                    self.close(price=profit)






class Turtle(bt.Strategy):

    def __init__(self,df):
        self.profit_percent = 1 # .3 # OG
        self.atr = bt.indicators.AverageTrueRange()

        self.time = self.datas[0].datetime



    def next(self):
        date = bt.utils.date.num2date(self.time[0])

        #use a try statement to avoid calculation errors on first candle
        try:
            '''if not currently in the market'''
            if not self.position:
                #only check trades if within timeframe
                if in_timeframe(date.time(),date.date()):
                    #get min and max prices in 55 candle period
                    maxPrice = max(self.data.high.get(ago=0,size=55))
                    minPrice = min(self.data.low.get(ago=0,size=55))

                    #buy conditions
                    if self.data.close <= minPrice and self.data.open > minPrice:
                        self.buy()


                    #sell conditions
                    elif self.data.close >= maxPrice and self.data.open < maxPrice:
                        self.sell()





            '''if a position is open'''
            if self.position:

                #close if end of trading week
                if date.date().weekday() == 5 and date.time() >= time(16) and date.time() <= time(16,6) or date.date().weekday() > 5:
                    self.close()

                
                '''check long'''
                if self.position.size > 0:

                    #take profit
                    if self.position.price + (self.position.price*(self.profit_percent/100)) > self.data.close:
                        self.close()

                    #stop loss
                    if self.position.price - (self.atr*2) < self.data.close:
                        self.close()



                '''check short'''
                if self.position.size < 0:

                    #take profit
                    if self.position.price - (self.position.price*(self.profit_percent/100)) < self.data.close:
                        self.close()

                    #stop loss
                    if self.position.price + (self.atr*2) > self.data.close:
                        self.close()
        except:
            pass
```
