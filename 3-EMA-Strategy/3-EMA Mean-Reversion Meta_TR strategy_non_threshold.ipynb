```python
#df[['Close', 'ema-short', 'ema-mid', 'ema-long']].plot()

def ma_constructor(data):
    df1 = data.copy()
    trend_lag  = 10

    df1['ema-short'] = df1['Close'].shift(1).ewm(alpha=(2/(9+1)), min_periods=10).mean()
    df1['ema-short-trend'] = df1['ema-short'] - df1['ema-short'].shift(trend_lag)

    df1['ema-mid'] = df1['Close'].shift(1).ewm(alpha=(2/(50+1)), min_periods=10).mean()
    df1['ema-mid-trend'] = df1['ema-mid'] - df1['ema-mid'].shift(trend_lag)
    df1['lower_threshold'] = df1['ema-mid'] * 0.99995
    df1['higher_threshold'] = df1['ema-mid'] * 1.00015

    df1['ema-long'] = df1['Close'].shift(1).ewm(alpha=(2/(100+1)), min_periods=10).mean()

    df1['ema-long-trend'] = df1['ema-long'] - df1['ema-long'].shift(trend_lag)



    display(df1.shape)

    apdict = mpf.make_addplot(df1[['ema-short', 'ema-mid', 'ema-long', 'lower_threshold', 'higher_threshold']])
    mpf.plot(df1,type='candle',volume=False, figsize=(20,10), addplot=apdict)
    display(df1.shape)


    df1.dropna(inplace=True)
    df1.reset_index(inplace=True)
    return df1

df_ma = ma_constructor(df)
# df_y_ma = ma_constructor(df_y)
# df_y_ma.rename(columns={'Datetime':'DateTime'}, inplace=True)

display(df_ma.head())
```

### enter rules
- All exponential moving average in same trend and space between them 
- candle breaking the 20-ema with an opposite trend
- next candle breaking the 20-ema favor to trend
- enter posision with stop-loss as the min of the previous candle if bullish, or max is the previous candle if bearish
- take profit as 1:1 in favor of the trend



```python

# 

def ma_3_strategy(data_ma):
    state = []
    hist = []
    stop_loss = []
    take_profit = []
    for i in data_ma.index:
        if len(hist) < 1:
            hist.append(0)

        elif len(state) < 1:

            #################### Entry rules #############

            # Evaluate if all three moving averages are have the same, significant trend.(bullish or bearish)


            # Bullish rules
            if (data_ma['ema-short-trend'][i] > 0) and (data_ma['ema-mid-trend'][i] > 0) and (data_ma['ema-long-trend'][i] > 0):


                    if (data_ma['Close'][i-1] < data_ma['ema-short'][i-1]) and (data_ma['Close'][i] > data_ma['ema-short'][i]):
                    
                        hist.append(1)
                        state.append(1)
                        stop_loss.append(data_ma['Close'][i])
                        take_p = data_ma['Close'][i] - data_ma['Open'][i] 
                        take_profit.append(data_ma['Close'][i] + (take_p * 1))

                    else:
                        hist.append(0)


            # Bearish rules
            elif (data_ma['ema-short-trend'][i] < 0) and (data_ma['ema-mid-trend'][i] < 0) and (data_ma['ema-long-trend'][i] < 0):




                    if (data_ma['Close'][i-1] > data_ma['ema-short'][i-1]) and (data_ma['Close'][i] < data_ma['ema-short'][i]):
      

                        hist.append(-1)
                        state.append(-1)
                        stop_loss.append(data_ma['Open'][i])
                        take_p = data_ma['Open'][i] - data_ma['Close'][i]
                        take_profit.append(data_ma['Close'][i] - (take_p * 1))

                    else:
                        hist.append(0)
      
            else:
                hist.append(0)

        elif state[-1] == 1:
            if (data_ma['Open'][i] <= stop_loss[-1]) or (data_ma['Open'][i] >= take_profit[-1]):
                print(data_ma['DateTime'][i])
                print(hist)
                print((data_ma['Open'][i] <= stop_loss[-1]))
                print((data_ma['Open'][i] >= take_profit[-1]))
                hist[-1] = 0
                hist.append(0)
                state.clear()
                stop_loss.clear()
                take_profit.clear()

            elif (data_ma['Close'][i] <= stop_loss[-1]) or (data_ma['Close'][i] >= take_profit[-1]):
                hist.append(0)
                state.clear()
                stop_loss.clear()
                take_profit.clear()

            else:
                hist.append(1)
                state.append(1)

        elif state[-1] == -1:

    #       Not enter position if stop loss or take profit thresholds were met before entering in position at the open
            if (data_ma['Open'][i] >= stop_loss[-1]) or (data_ma['Open'][i] <= take_profit[-1]):
                
                hist[-1] = 0
                hist.append(0)
                state.clear()
                stop_loss.clear()
                take_profit.clear()

            # Close position if stop loss or take profit thresholds were met at the close
            elif (data_ma['Close'][i] >= stop_loss[-1]) or (data_ma['Close'][i] <= take_profit[-1]):
                hist.append(0)
                state.clear()
                stop_loss.clear()
                take_profit.clear()

            else:
                hist.append(-1)
                state.append(-1)
    print(hist)  
    return hist



df_strat = ma_3_strategy(df_ma)            
    
```

```python
def gen_stats(data, returns, freq=None, window=None, strategy=False):
    idx = pd.date_range('2022-05-02', '2022-10-07', freq="d")
    nyse = mcal.get_calendar('NYSE')
    holidays_nyse = nyse.holidays().holidays
    data = data.copy()
    data = data.resample('d').sum()
    data.drop_duplicates(inplace=True)
    data = data.reindex(idx)
    print(data.shape)
    data = data[data.index.dayofweek < 5]
    data = data [~data.index.isin(holidays_nyse)]
    data.fillna(0, inplace=True)
    fract_trade = (data[data[returns] != 0.000000].shape[0])/data.shape[0]
    target_ret = data[returns]
    pfstats = pd.DataFrame(columns=['Strategy'])
    # Mean / std. dev. of return
    ret_m = target_ret.mean()
    ret_s = target_ret.std()
    if freq == 'daily':
        ret_ann = ret_m*252
        ret_vol = ret_s*np.sqrt(252)
    if freq == 'weekly':
        ret_ann = (1+ret_m)*52-1
        ret_vol = ret_s*np.sqrt(52)
    pfstats.loc['Ann. Return'] = str(round(ret_ann * 100, 1)) + '%'
    pfstats.loc['Ann. Volatility'] = str(round(ret_vol * 100, 1)) + '%'
    # Sharpe
    sharpe = ret_ann/ret_vol
    pfstats.loc['Sharpe Ratio (ann.)'] = sharpe.round(2)
    # Maximum Drawdown rolling
    cum_ret = 1000*(1+target_ret).cumprod()-1
    max_dd = cum_ret/cum_ret.rolling(window).max()-1
    max_dd = max_dd.min()
    if freq == 'weekly':
        max_dd_string = str(window)+'-W'
    else:
        max_dd_string = str(window)+'-D'
    pfstats.loc['Max. DD ('+max_dd_string+' rolling)'] = str(round(max_dd, 1))
    # Skewness
    ret_skew = target_ret.skew()
    if freq == 'daily':
        ret_skew_ann = ret_skew/np.sqrt(252)
    pfstats.loc['Skew (ann.)'] = ret_skew_ann.round(2)
    # Fraction of days traded
    pfstats.loc['Days Traded'] = str(int(round(fract_trade * 100, 2))) + '%'
    return pfstats
```
