```python
import pandas as pd 
import os

ticker = ['NQ']
path = r'C:\Users\ianaz\Documents\Upwork projects\Miracle_fund\3-EMA-Strategy'


df_list = []
for file in os.listdir(path):
    df_list.append(os.path.join(path, file))
    
ticker_path = [i for i in df_list if '2Min' in i]

ticker_path


QQQ_whole = pd.DataFrame()
for i in ticker_path:
    df_temp = pd.read_csv(i)
    QQQ_whole = QQQ_whole.append(df_temp)
    display(df_temp.head())

QQQ_whole.drop(columns='Unnamed: 0', inplace=True)

QQQ_whole.set_index('DateTime', inplace=True)
QQQ_whole.index = pd.to_datetime(QQQ_whole.index)
QQQ_whole.drop_duplicates(inplace=True)

QQQ_whole.to_csv('NQ_whole_2Min.csv')
```
